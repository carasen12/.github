For more information on how to use the Inference Providers, please refer to our [documentation and guides](https://huggingface.co/docs/inference-providers/en/index).

````python
import os
os.environ['HF_TOKEN'] = 'YOUR_TOKEN_HERE'
````

````python
import os
from huggingface_hub import InferenceClient

client = InferenceClient(
    provider="auto",
    api_key=os.environ["HF_TOKEN"],
)

completion = client.chat.completions.create(
    model="google/DiarizationLM-13b-Fisher-v1",
    messages="\"Can you please let us know more details about your \"",
)

print(completion.choices[0].message)
````

## Experiment with local LLMs via LM Studio for rapid prototyping

### Subtask:
Outline steps to set up and connect local Large Language Models (LLMs) served by LM Studio to a Colab notebook for local inference, offering flexibility and rapid experimentation with different models.


4. **Connect from Colab (using `ollama` or `openai` libraries)**:
   To interact with the LM Studio server from your Colab notebook, you will typically use an OpenAI-compatible client library, as LM Studio provides an OpenAI-compatible API endpoint.
   - **Install `openai` library** in your Colab notebook: `!pip install openai`
   - **Configure the client** to point to your local LM Studio server. You will need to make sure your Colab environment can reach your local machine's IP address. This often requires port forwarding or using a tunnel (like ngrok) if you are running Colab on a cloud instance and LM Studio on a truly local machine, or if there are network restrictions. For a simpler setup, if Colab is running on a local runtime, `http://localhost:1234` might work directly.

   ```python
   from openai import OpenAI

   # Replace with your LM Studio server address and port
   # If Colab is on a cloud instance, you'll need ngrok or similar to expose localhost
   # Example with ngrok: client = OpenAI(base_url="http://<ngrok-id>.ngrok.io/v1", api_key="not-needed")
   # For a local Colab runtime, or if properly networked:
   client = OpenAI(base_url="http://localhost:1234/v1", api_key="not-needed")

   # Optional: Test the connection
   try:
       models = client.models.list()
       print("Successfully connected to LM Studio server!")
       print("Available models:", [model.id for model in models.data])
   except Exception as e:
       print(f"Failed to connect or retrieve models: {e}")
       print("Ensure LM Studio server is running and accessible from Colab (check networking/port forwarding).")
   ```
